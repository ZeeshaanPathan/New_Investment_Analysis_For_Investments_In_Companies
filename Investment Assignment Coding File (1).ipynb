# Merging on the commonly exsiting column to create a Master file.

For doing this we have observed that the data set case is  different in both the files of Companies and rounds2. Hence converting them to similar case upper/ lower so that we can merge the file on a common column.

In our case , let's convert the entries to upper case.

Upper case entries of rounds2 file :

```python
rounds2['company_permalink'] = rounds2['company_permalink'].apply (lambda x : x.upper())
```

*Converting the string back " raised_amount_usd" to float for calculations*

Upper case entries in the companies file / dataframe:

```python
companies ['permalink'] = companies['permalink'].apply(lambda x: x.upper())
```

Creating a master_frame from the above dataframes

```python
master_frame = rounds2.merge(companies , left_on = 'company_permalink', right_on = 'permalink')
master_frame.head()
```

# Checkpoint 4: Sector Analysis 1


Creating a data frame of the Top 3 English Speaking countries for investments

```python
dfusa = master_frame.loc[(master_frame["country_code"] == "USA")  , :]
dfusa
```

```python
dfind = master_frame.loc[(master_frame["country_code"] == "IND")  , :]
dfind
```

```python
dfgbr = master_frame.loc[(master_frame["country_code"] == "GBR")  , :]
dfgbr
```

Creating Master file of investments in top 3 countries

```python
master_frame_top3 = pd.concat([dfusa, dfind, dfgbr], axis = 0)
master_frame_top3
```

```python
mf1 = master_frame_top3 ["category_list"].str.split("|", n = 1, expand = True) 
mf1
```

```python
master_frame_top3 ["category_list"]= mf1[0] 
master_frame_top3.head()
```

Importing the mapping file.

```python
mapping = pd.read_csv ("F:\\Studies\\AI and ML\\Investment Assignment Files\\mapping.csv", encoding = "ISO-8859-1")
mapping.head()
```

Creating the final mapping file

```python
mf2 = pd.merge (master_frame_top3 , mapping , how= 'inner' , on = 'category_list')
mf2.head()
```

```python
mff = mf2.loc [(mf2['raised_amount_usd'] > 5000000) & (mf2['raised_amount_usd'] < 15000000) , :]
mff.info()
```
